```python
import pandas as pd
import glob
```

```python
path = '/N/slate/maahutch/WT_query3'
all_files = glob.glob(path + "/*.csv")
```

```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, sep = '\t', header = 0, quotechar='"', error_bad_lines=False)
    li.append(df)
    
frame = pd.concat(li, axis=0, ignore_index = True)
```
